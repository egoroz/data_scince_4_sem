```python
wine_data = pd.read_csv("wine.data", header=None)
# print(wine_data)

wine_type_1 = wine_data[wine_data[0] == 1]

columns_of_interest = [3, 7, 8, 10]  # `Alcalinity of ash`, `Nonflavanoid phenols`, `Proanthocyanins` и `Hue`
wine_type_1_subset = wine_type_1[columns_of_interest]

sample_means = wine_type_1_subset.mean()
sample_std_devs = wine_type_1_subset.std()

alpha = 0.95

n = len(wine_type_1_subset)

quantile_normal = sps.norm.ppf((1 + alpha) / 2)
error_norm = quantile_normal/np.sqrt(n) * np.sqrt(sample_std_devs)
# print(sample_means - error_norm)

quantiles_t = sps.t.ppf((1 + alpha) / 2, df=n-1)
error_t = quantiles_t/np.sqrt(n-1) * np.sqrt(sample_std_devs)

chi2_lower = sps.chi2.ppf((1 - alpha) / 2, df = n - 1)
chi2_upper = sps.chi2.ppf((1 + alpha) / 2, df = n - 1)

error_lower = chi2_lower/np.sqrt(n-1) * np.sqrt(sample_std_devs)
error_upper = chi2_upper/np.sqrt(n-1) * np.sqrt(sample_std_devs)

confidence_intervals = pd.DataFrame({
    "Column": wine_type_1_subset.columns,
    "Sample Mean": sample_means.round(4),
    "Norm lower": (sample_means - error_norm),
    "Norm upper": (sample_means + error_norm),
    "chi2 lower": (sample_means - error_lower),
    "chi2 upper": (sample_means + error_upper),
    "student lower": (sample_means - error_t),
    "student upper": (sample_means + error_t),
})

print(confidence_intervals)

```

### Функция правдоподобия
Дана параметрическая модель и 3 выборки, состоящие из 2-3 наблюдений. Для удобства, выборки представлены в виде python-кода — каждая выборка записана как список ее элементов; множество выборок представлено как список списков, соответствующих выборкам из множества. Нужно для каждой выборки построить график функции правдоподобия.

*a).* Параметрическая модель $\mathcal{N}(\theta, 1)$, выборки: `[[-1, 1], [-5, 5], [-1, 5]]`

*b).* Параметрическая модель $Exp(\theta)$, выборки: `[[1, 2], [0.1, 1], [1, 10]]`

*c).* Параметрическая модель $U[0, \theta]$, выборки: `[[0.2, 0.8], [0.5, 1], [0.5, 1.3]]`

*e).* Параметрическая модель $Pois(\theta)$, выборки: `[[0, 1], [0, 10], [5, 10]]`

*f).* Параметрическая модель $Сauchy(\theta)$, где $\theta$ — параметр сдвига, выборки: `[[-0.5, 0.5], [-2, 2], [-4, 0, 4]]`

Выполнить задание, не создавая много кода, поможет следующая функция.

Допишите ее.

```python
def draw_likelihood(density_function, grid, samples, label):
    """Изображает график функции правдоподобия для каждой из 3 выборок.

    Аргументы:
    density_function --- функция, считающая плотность (обычную или дискретную).
        На вход данная функция должна принимать массив размера (1, len_sample)
        и возвращать массив размера (len_grid, len_sample).
    grid --- массив размера (len_grid, 1) --- сетка для построения графика;
    samples --- три выборки;
    label --- latex-код параметрической модели.
    """
    assert len(samples) == 3, "Число выборок не равно 3."

    plt.figure(figsize=(18, 5))
    for i, sample in enumerate(samples):
        sample = np.array(sample)[np.newaxis, :]
        likelihood = density_function(sample).prod(axis=1)

        plt.subplot(1, 3, i+1)
        plt.plot(grid, likelihood)
        plt.xlabel('$\\theta$', fontsize=16)
        plt.grid(ls=':')
        plt.title(label + ', sample=' + str(sample), fontsize=16)
    plt.show()
```

Выполните:

```python
# def normal_density(theta, sample=):
#     return sps.norm.pdf(sample, loc=theta, scale=1)

def exponential_density(theta, sample):
    return sps.expon.pdf(sample, scale=1/theta)

def uniform_density(theta, sample):
    return sps.uniform.pdf(sample, loc=0, scale=theta)

def binomial_density(theta, sample):
    return sps.binom.pmf(sample, n=5, p=theta)

def poisson_density(theta, sample):
    return sps.poisson.pmf(sample, mu=theta)

def cauchy_density(theta, sample):
    return sps.cauchy.pdf(sample, loc=theta, scale=1)

samples_a = [[-1, 1], [-5, 5], [-1, 5]]
samples_b = [[1, 2], [0.1, 1], [1, 10]]
samples_c = [[0.2, 0.8], [0.5, 1], [0.5, 1.3]]
samples_d = [[0, 1], [5, 5], [0, 5]]
samples_e = [[0, 1], [0, 10], [5, 10]]
samples_f = [[-0.5, 0.5], [-2, 2], [-4, 0, 4]]

grid = np.linspace(-10, 20, 1000).reshape((-1, 1))
draw_likelihood(sps.norm(loc=grid).pdf, grid, samples_a, label='$\\mathcal{N}(\\theta, 1)$')
draw_likelihood(sps.expon(scale=grid).pdf, grid, samples_b, label='$Exp(\\theta)$')
draw_likelihood(sps.uniform(scale=grid).pdf, grid, samples_c, label='$U[0, \\theta]$')
draw_likelihood(sps.binom(5,grid).pmf, grid, samples_d, label='$Bin(5, \\theta)$')
draw_likelihood(sps.poisson(grid).pmf, grid, samples_e, label='$Pois(\\theta)$')
draw_likelihood(sps.cauchy(loc=grid).pdf, grid, samples_f, label='$Cauchy(\\theta)$')
```

**Ответ на вопрос и описание метода решения проблемы:**


Метод .pdf(sample).prod() вычисляет плотность вероятности для каждого элемента выборки и затем перемножает все значения. Но для большой выборки, произведение плотностей вероятности может стать очень маленьким числом, близким к нулю (в нашем случае = 0).

Можем использовать логарифмическую функцию правдоподобия. То есть будем вычислять плотность для каждого элемента выборки, а затем суммировать

```python
sample = sps.norm.rvs(size=10**5)

log_likelihood = sps.norm.logpdf(sample).sum()

print(f"Логарифмическая функция правдоподобия: {log_likelihood:.4e}")
```

## Задача 6
Асимпотические доверительные интервалы ничего не могут гарантировать на малых размерах выборки.
В этой задаче вам предстоит иллюстрировать этот факт, посчитав реальный уровень доверия для интервалов.

**Реальный уровень доверия (оценка доли покрытия интервалом)**  - доля случаев попадания истинного значения параметра в доверительный интервал.

*Пример подсчета:*

Допустим, вы хотите оценить реальный уровень доверия интервала для $a$, если $X \sim \mathcal{N}(a, \sigma^2)$

- Фиксируете истинные $(a, \sigma^2)$, для которых будете делать оценку
- Генерируете $B$ выборок из $\mathcal{N}(a, \sigma^2)$ с зафиксированными параметрами
- По каждой выборке получаете АДИ для $a$ 
- Считаете долю случаев, когда истинное $a$ попадает в интервал

Эта доля и будет оценкой доли покрытия интервала.

**Важно:** вы симулируете реальную ситуацию, когда вы не знаете ни $a$, ни $\sigma$, поэтому в формулах для АДИ их **использовать нельзя**!

**Важно:** при такой оценке реального уровня доверия вы используете метод Монте-Карло. Погрешность этого метода составляет $\sim \frac{1}{\sqrt{n}}$, где $n$ - количество выборок, по которым осуществляется оценка.

Постройте график зависимости реального уровня доверия интервала от размера выборки. График начинайте с $n=2$.

Для этого вам нужно провести описанную выше процедуру для подвыборок разных размеров. Для подсчета интервалов по префиксам используйте функцию `np.cumsum`. Для того чтобы не запутаться, разносите подсчет разных величин, входящих в формулу, по разным строчкам. 

```python
theta = 0
sample_count = 10000
max_sample_size = 300

sample_sizes = np.arange(2, max_sample_size + 1)
real_cov_rates = np.zeros(max_sample_size - 1)

samples = sps.norm.rvs(loc=theta, scale=1, size=(sample_count, max_sample_size))

cum_means = np.cumsum(samples, axis=1) / np.arange(1, max_sample_size + 1)
cum_std_devs = np.sqrt(np.cumsum((samples - cum_means) ** 2, axis=1) / np.arange(1, max_sample_size + 1))

# Рассчитываем доверительные интервалы для каждого размера выборки
for i, n in enumerate(sample_sizes):
    lower_bound = cum_means[:, i] - cum_std_devs[:, i] / np.sqrt(n) * z
    upper_bound = cum_means[:, i] + cum_std_devs[:, i] / np.sqrt(n) * z
    true_theta = (lower_bound <= theta) & (theta <= upper_bound)
    real_cov_rates[i] = np.mean(true_theta)

# Построим график реальных уровней доверия
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, real_cov_rates, marker='o', linestyle='-', color='b', label='Асимптотический интервал')
plt.axhline(y=alpha, color='r', linestyle='--', label='Теоретический уровень доверия')
plt.xlabel('Размер выборки')
plt.ylabel('Реальный уровень доверия')
plt.title('Реальный уровень доверия vs. Размер выборки')
plt.legend()
plt.grid(True)
plt.show()

# Выведем реальный уровень доверия для самого большого размера выборки
print(f"Реальный уровень доверия для n={max_sample_size}: {real_cov_rates[-1]:.4f}")
```

Сделайте выводы.

реальный уровень доверия начинает сильно отличаться от теоретического при малых размерах выборки. При увеличении выборки уровень доверия стремится к теоретическому. Метод Монте-Карло дает хорошую оценку на колличество выборок (точность 0.01 выполняется)

### Случай ТДИ

На лекции вы получали формулу для точного доверительного интервала для $a$ в нормальной модели

Точный доверительный интервал: $\theta \in \bigg(\overline{X} - \frac{S}{\sqrt{n-1}}T_{n-1,(1+\alpha)/2}, \overline{X} + \frac{S}{\sqrt{n-1}}T_{n-1,(1+\alpha)/2} \bigg)$

**Вопрос:** чем этот интервал лучше предыдущего?

**Ваш ответ:** 

Постройте график реального уровня доверия интервала от размера выборки для этого вида интервала. График начинайте с $n=2$. Сравните его с предыдущим. 

```python
import numpy as np
import matplotlib.pyplot as plt

theta = 0
sample_count = 10000
max_sample_size = 300

real_cov_rates_exact = np.zeros(max_sample_size - 1)

samples = sps.norm.rvs(loc=theta, scale=1, size=(sample_count, max_sample_size))

cum_means_exact = np.cumsum(samples, axis=1) / np.arange(1, max_sample_size + 1)
cum_std_devs_exact = np.sqrt(np.cumsum((samples - cum_means_exact) ** 2, axis=1) / np.arange(1, max_sample_size + 1))

for i, n in enumerate(sample_sizes):
    lower_bound_exact = cum_means_exact[:, i] - cum_std_devs_exact[:, i] / np.sqrt(n - 1) * sps.t.ppf((1 + alpha) / 2, df=n - 1)
    upper_bound_exact = cum_means_exact[:, i] + cum_std_devs_exact[:, i] / np.sqrt(n - 1) * sps.t.ppf((1 + alpha) / 2, df=n - 1)
    true_theta_exact = (lower_bound_exact <= theta) & (theta <= upper_bound_exact)
    real_cov_rates_exact[i] = np.mean(true_theta_exact)

plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, real_cov_rates, marker='o', linestyle='-', color='b', label='Асимптотический интервал')
plt.plot(sample_sizes, real_cov_rates_exact, marker='o', linestyle='-', color='g', label='Точный интервал')
plt.axhline(y=alpha, color='r', linestyle='--', label='Теоретический уровень доверия')
plt.xlabel('Размер выборки')
plt.ylabel('Реальный уровень доверия')
plt.title('Сравнение реальных уровней доверия (точный vs. асимптотический)')
plt.legend()
plt.grid(True)
plt.show()

print(f"Реальный уровень доверия для n={max_sample_size} (точный интервал): {real_cov_rates_exact[-1]:.4f}")

```
