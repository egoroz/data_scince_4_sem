## Задача 1

Проверьте, отличается ли средний балл студентов курса Phystech@DataScience за лекгкую часть задания 3 и 5 весной 2023.

Данные возьмите из [таблицы оценок](https://docs.google.com/spreadsheets/d/1GSw2_vAMQ4XZcAf8Y_99CkJ1oi4_LrHyWIS-DCdFE0Y/edit?usp=sharing). Не забудьте визуализировать данные с помощью ядерной оценки плотности (вам поможет функция `sns.kdeplot`) и оценить сдвиг.



```python
#Подготовка данных для работы
Columns = ["Задание 3", "Задание 5"]
df = pd.read_csv(
    "ex.csv")[Columns].drop([0])

for col in Columns:
    df[col] = df[col].str.replace(",", '.')
    df[col] = df[col].to_numpy().astype(
        np.float64)
#Рисую график
colors = ["red", "blue"]
fig=plt.figure(figsize=(5, 5), dpi=150)

for color, (col, data) in zip(colors, df.items()):
    sns.kdeplot(data,
                color=color,
                alpha=0.7)
    
    plt.axvline(np.mean(df[col]),
            linewidth=0.5,
            label='Среднее ' + col,
            color=color,
            linestyle='--')
plt.title("Ядерная оценка плотности")
plt.legend()

#Оценю сдвиг
def count_conf(X: np.ndarray, Y: np.ndarray, alpha: float = 0.05):
    S_x2 = np.var(X)
    S_y2 = np.var(Y)
    z = sps.t(df=len(X) + len(Y) - 2).ppf(1 - alpha / 2)
    center = np.mean(X) - np.mean(Y)
    result = {"center": center,
              "l": center - z * np.sqrt(S_x2 / len(X) + S_y2 / len(Y)),
              "r":  center + z * np.sqrt(S_x2 / len(X) + S_y2 / len(Y))}
    return result

conf = count_conf(df['Задание 3'], df['Задание 5'])
print(f"Оценка сдвига: {conf['center']}")
print(f"доверительный интервал: [{conf['l']} ; {conf['r']}]")

```

Сделайте общий вывод по задаче.

**Вывод:**
Среднее за пятое задание меньше среднего за третье. Т.к. Ноль в доверительный интервал не попадает, из этого следует, что средние значение балла третьего дз больше среднего значения за пятое, т.е. отвергается гипотеза равенства средних

## Задача 2
Рассмотрите таблицу с оценками по какому-либо курсу с несколькими семинарскими группами. Выберите свою группу и еще какую-нибудь. Проверьте гипотезу о том, что средний балл не отличается между семинарскими группами. Не забывайте про визуализацию.

Сделайте выводы.

```python
#Подготовка данных для работы: колонки - номер группы, строчки - номер человека.
df = pd.read_csv(
    "ex_2.csv", sep=';',)

#Рисую график
colors = ["red", "blue",'green', 'black', 'pink', 'grey', 'yellow']
fig = plt.figure(figsize=(8, 5), dpi=150)

for color, (col, data) in zip(colors, df.items()):
    sns.kdeplot(data,
                color=color,
                alpha=0.7)

    plt.axvline(np.mean(df[col]),
                linewidth=0.5,
                label='Среднее ' + col + ' группы',
                color=color,
                linestyle='--')
plt.title("Ядерная оценка плотности")
plt.legend()

#Оценю сдвиг


def count_conf(X: np.ndarray, Y: np.ndarray, alpha: float = 0.05):
    S_x2 = np.var(X)
    S_y2 = np.var(Y)
    z = sps.t(df=len(X) + len(Y) - 2).ppf(1 - alpha / 2)
    center = np.mean(X) - np.mean(Y)
    result = {"center": center,
              "l": center - z * np.sqrt(S_x2 / len(X) + S_y2 / len(Y)),
              "r":  center + z * np.sqrt(S_x2 / len(X) + S_y2 / len(Y))}
    return result

diny = 0
undiny = 0
for i in range(int(df.shape[1])):
    for j in range(df.shape[1]-i-1):
        conf = count_conf(df.iloc[i], df.iloc[i+1+j])
        print(f"Группа {i+1} и группа {i+j+2}")
        print(f"Оценка сдвига: {conf['center']}")
        print(f"доверительный интервал: [{conf['l']} ; {conf['r']}]")
        if (conf['l'] > 0) or (conf['r']<0):
            print(f"Ноль не принадлежит доверительному интервалу гипотеза о  равенстве средних у групп {i+1}, {i+j+2} отвергается")
            diny+=1
        else:
            undiny += 1
            print(
                f"Ноль принадлежит доверительному интервалу гипотеза о  равенстве средних у групп {i+1}, {i+j+2} не отвергается")
print(f"отвергнуто: {diny}, не отвергнуто: {undiny}")
```

**Вывод:**
Как можно увидеть из напечатанного гипотеза о равентстве средних не отвергается ни в одном случае.

```python
path = 'ex_3.csv'
data = pd.read_csv(path)
types = data['id'].unique()
data.sample(10)
```

Постройте гистограммы признаков для каждого вида частиц.

```python
Columns = ['p', 'theta', 'beta', 'nphe', 'ein', 'eout']
#В словарь словарей limits вставляем ограничения оси x
limits = {
    211:
    {
        'p': (0, 4),
        'theta': (0, 2),
        'beta': (0.8, 1.1),
        'nphe': (-0.2, 5),
        'ein': (-0.1, 0.6),
        'eout': (-0.1, 0.6)
    },
    2212:
    {
        'p': (0, 4),
        'theta': (0, 1.25),
        'beta': (0.2, 1.1),
        'nphe': (-0.2, 5),
        'ein': (-0.1, 0.6),
        'eout': (-0.1, 0.6)
    },
    321:
    {
        'p': (0, 4),
        'theta': (0, 2),
        'beta': (0.5, 1.2),
        'nphe': (-0.2, 5),
        'ein': (-0.1, 0.6),
        'eout': (-0.1, 0.6)
    },
    -11:
    {
        'p': (0, 3.5),
        'theta': (0, 2),
        'beta': (0.9, 1.1),
        'nphe': (-0.2, 5),
        'ein': (-0.1, 0.6),
        'eout': (-0.1, 0.06)
    }
}

for type in types:
    fig, ax = plt.subplots(2, 3, figsize=(16, 9), dpi=200)
    fig.suptitle(f'Тип частицы: id = {type}')
    for ind, (label, column) in enumerate(data[data.id == type][Columns].items()):
        X, Y = ind//3, ind % 3
        mean= column.mean()
        ax[X,Y].hist(column,
                            color='blue',
                            bins=200,
                            density=True,
                            alpha=0.7,
                            label='Распределение')
        ax[X, Y].axvline(mean,
                               linestyle='--',
                               color='red',
                               label=f'Среднее значение $ = {mean:.3f}$')

        ax[X, Y].set_xlim(limits[type][label])
        ax[X, Y].set_title(label)
        ax[X, Y].legend()
        ax[X, Y].set_xlabel("$x$")
        ax[X, Y].set_ylabel("$p(x)$")
```
