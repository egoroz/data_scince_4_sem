Тип вина указан в первом столбце. Для работы с табличными данными используйте библиотеку `pandas`.

Постройте доверительные интервалы для параметров сдвига каждого из столбцов, предполагая, что столбцы имеют нормальное распределение.
Требуется построить: 
 * асимтотические доверительные интервалы при помощи центральной предельной теоремы;
 * точные неасимптотические при помощи распределений хи-квадрат, Стьюдента.

 
 Запишите их в виде таблицы.

```python
df = pd.read_csv('wine.data')
df = df[df['1']==1].drop(columns='1')
columns = ['Alcohol',
          'Malic acid',
          'Ash',
          'Alcalinity of ash',
          'Magnesium',
          'Total phenols',
          'Flavanoids',
          'Nonflavanoid phenols',
          'Proanthocyanins',
          'Color intensity',
          'Hue',
          'OD280/OD315 of diluted wines',
          'Proline']
add_data = [float(np.array(df.columns)[i])for i in range(len(df.columns))]
df.columns = columns

df.drop(columns=['Alcohol',
                 'Malic acid',
                 'Ash',
                 
                 'Magnesium',
                 'Total phenols',
                 'Flavanoids',
                 
        
                 'Color intensity',
               
                 'OD280/OD315 of diluted wines',
                 'Proline'], inplace=True)

alpha = 0.05

df.loc[len(df.index)]= np.array([df.iloc[:, i].std()*sps.norm.ppf(1-alpha/2) /
          (len(df.iloc[:,0]))**0.5 for i in range(len(df.columns))]).round(2)


df.loc[len(df.index)] = np.array([df.iloc[:, i].std()*sps.f.ppf(q=1-alpha/2, dfd=len(df.iloc[:, 0]), dfn=1) /
          (len(df.iloc[:, 0])-1)**0.5 for i in range(len(df.columns))]).round(2)
df.drop([i for i in range(len(df.iloc[:-2, 0]))], inplace=True)
df.rename({58: 'АДИ', 59: 'ТДИ'}, axis='index')
```

Сделайте выводы по полученной таблице.



**Вывод:**

Не достатчно экспериментов для использования АДИ, нужен ТДИ, т.к. заметное рассхождение в значениях доверительного интервала

## Задача 3

Постройте плотность стандартного **нормального распределения** и стандартного **распределения Коши**. Видим, что распределение Коши обладает более тяжелыми хвостами, нежели нормальное.

```python
grid = np.linspace(-7, 7, 1000)

plt.figure(figsize=(10, 5))
plt.plot(grid, sps.cauchy.pdf(grid), color=red, label='Коши', lw=3)
plt.plot(grid, sps.norm.pdf(grid), color=blue, 
         label='$\mathcal{N}(0, 1)$', lw=3)
plt.title('Плотности распределений')
plt.legend();
```

**Вывод:**
