**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

Для ускорения расчетов можете использовать broadcasting (не является обязательным требованием). Учите, что в таком случае внутри функции у вас могут получиться очень большие матрицы, из-за которых может кончиться оперативная память. 

```python
def bootstrap(sample, B=1000): 
    '''
    Считает бутстрепные оценки для исходной выборки

    :param sample: исходная выборка
    :param B: количество бутстрепных 
    :return bootstrap_estimations: оценки по бутстрепным выборкам
    '''   
    
    B = len(sample)
    bootstrap_estimations = np.random.choice(sample, size=(B, B), replace=True)
    bootstrap_estimations = np.sort(np.mean(bootstrap_estimations, axis=1))

    return bootstrap_estimations
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. 

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу нормального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   


    v = np.var(bootstrap_estimations)
    
    z = sps.norm.ppf((1 + alpha) / 2)
    
    left = theta - z * v**0.5
    right = theta + z * v**0.5

    return round(left, 2), round(right, 2)


    
def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу центрального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   

    B = len(bootstrap_estimations)

    
    left = 2*theta -  bootstrap_estimations[math.ceil(B*(1+alpha)*0.5)-1]
    right = 2*theta - bootstrap_estimations[int(B*(1-alpha)*0.5)-1]
    
    
    return round(left, 2), round(right, 2)


    
def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу квантильного бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    
    B = len(bootstrap_estimations)
    
    left =  bootstrap_estimations[int(B*(1-alpha)/2)-1]
    right =  bootstrap_estimations[math.ceil(B*(1+alpha)/2)-1]

    return round(left, 2), round(right, 2)
   
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

```python
dat = pd.DataFrame(index = ['normal', 'central', 'quantile', 'theta', 'interval_length'], columns=df.columns)

for col in df.columns:
    # Получим бутстрепные оценки
    bootstrap_estimations = bootstrap(df[col])

    # Вычислим оценку параметра
    theta = round(np.mean(df[col]), 2)

    # Построим доверительные интервалы
    dat.loc['normal', col] = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
    dat.loc['central', col] = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
    dat.loc['quantile', col] = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)
    
    dat.loc['theta', col] = theta
    dat.loc['interval_length', col] = [round(dat.loc['normal', col][1] - dat.loc['normal', col][0], 2),
                                       round(dat.loc['central', col][1] - dat.loc['central', col][0], 2),
                                       round(dat.loc['quantile', col][1] - dat.loc['quantile', col][0], 2)]

dat
```

**Вывод:** бутстрепные интервалы для конкретного признака не сильно отличаются друг от друга.

Визуализируйте бутстрепные интервалы для каждого признака. Для этого сгенерируйте выборку $X_1, ... X_{N}, N = 100$ и постройте график доверительных интервалов уровня доверия $0.95$, вычисленных для всех подвыборок размера $n$ вида $X_1, ... X_n$, $1 \le n \le 100$, используя написанную ниже функцию

```python
def draw_confidence_interval(
    left,
    right,
    estimation=None,
    sample=None, 
    ylim=(None, None), 
    estim_label = '',
    sample_label='',
    color=("green", "red", "blue"),
    interval_label=None,
    title=''
):
    '''
    Рисует доверительный интервал и оценку в зависимости от размера выборки.
    
    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param estim_label: подпись к оценке
    :param sample_label: подпись к выборке
    :param color: цвет, которым будет отображен доверительный интервал
    '''
    lin = ["normal confidence interval", "central confidence interval", "quantile confidence interval"]

    
    plt.figure(figsize=(10, 6))
    for i in range(len(left)):
        plt.fill_between(range(1, len(sample) + 1), left[i], right[i], color=color[i], alpha=0.08, label=lin[i])
        plt.plot(range(1, len(sample) + 1), estimation[i], color=color[i], label=f"sample min for {lin[i]}")
    for i in range(0, len(sample)):
        plt.scatter((i+1) * np.ones(len(sample[i])), sample[i], color='blue', alpha=0.005)
    plt.xlabel('Sample Size (n)')
    plt.ylabel('Value')
    plt.title(f'Confidence Intervals for {title}')
    plt.ylim(ylim)
    plt.legend()
    plt.show()
```

**Решение:**

```python
n_values = np.arange(1, 101)

sample_size = 100


np.random.seed(207)


col = ["E1", "E2", "M"]
lin = ["normal", "central", "quantile"]



for i,c in enumerate(col):
    left_intervals = [[], [], []]
    right_intervals = [[], [], []]
    estimations = [[], [], []]
    samples = []

    for n in range (2, sample_size):
        data = df.head(n)
        samples.append(data[c].to_numpy())
  
        
        bootstrap_estimations = bootstrap(data[col[0]])
        theta = round(np.mean(data[col[0]]), 2)
        estimations[0].append(theta)
        interval = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
        left_intervals[0].append(interval[0])
        right_intervals[0].append(interval[1])
        
        ootstrap_estimations = bootstrap(data[col[1]])
        theta = round(np.mean(data[col[1]]), 2)
        estimations[1].append(theta)
        interval = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
        left_intervals[1].append(interval[0])
        right_intervals[1].append(interval[1])   
        
        ootstrap_estimations = bootstrap(data[col[2]])
        theta = round(np.mean(data[col[2]]), 2)
        estimations[2].append(theta)
        interval = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)
        left_intervals[2].append(interval[0])
        right_intervals[2].append(interval[1])   
    
    
    draw_confidence_interval(
        left_intervals,
        right_intervals,
        estimation=estimations,
        sample=samples,
        ylim=None,
        estim_label='Sample Mean',
        sample_label='Sample',
        color=('green', "blue", "red"),
        interval_label=f'Feature {i+1} Confidence Interval',
        title=c
    )
```

**Вывод:** 

Нормальный бутстрепный доверительный интервал: оценивает с использованием нормального распределения.

**2.** Чему равна асимптотическая дисперсия $\sigma_{\alpha}^2$ усеченного среднего $\overline{X}_{\alpha}$, где $0 < \alpha < 1/2$, для распределения Коши? Помочь в ее вычислении может теорема, упомянутая на лекциях. Посчитайте асимптотическую дисперсию аналитически. Постройте график зависимости асимптотической дисперсии $\sigma_{\alpha}^2$ усеченного среднего $\overline{X}_{\alpha}$, где $0 < \alpha < 1/2$, от $\alpha$.

Настройте видимую область графика по оси *y* так, чтобы четко была отображена информативная часть графика. Отметьте на графике минимум функции.

СМ. ПРИКРЕПЛЕННЫЙ ФАЙЛ.

**Решение:**

```python

def z(q):
    return sps.cauchy.ppf(q)


def asymptotic_variance(alpha):
    return (2 / (1 - 2 * alpha)**2) * ((1/np.pi)*z(1-alpha) - (1/np.pi)*np.arctan(z(1 - alpha)) + alpha*(z(1-alpha))**2)


alpha_values = np.linspace(0.01, 0.99, 100000)

# Вычисляем асимптотическую дисперсию для каждого alpha
variance_values = asymptotic_variance(alpha_values)

# Находим минимальное значение дисперсии и соответствующее alpha
min_variance = np.min(variance_values)
optimal_alpha = alpha_values[np.argmin(variance_values)]


plt.figure(figsize=(10, 10))
plt.plot(alpha_values, variance_values, label='Асимптотическая дисперсия')
plt.scatter(optimal_alpha, min_variance, color='red', marker='o', label='Минимум дисперсии')
plt.xlabel('Alpha (Уровень усечения)')
plt.ylabel('Асимптотическая дисперсия')
plt.title('Асимптотическая дисперсия усеченного среднего для распределения Коши')
plt.legend()
plt.grid(True)
plt.ylim(0, 10)  
plt.show()


print(f"Минимальная асимптотическая дисперсия: {min_variance:.6f} (при alpha = {optimal_alpha:.6f})")
```

## Задача 6
Асимпотические доверительные интервалы ничего не могут гарантировать на малых размерах выборки.
В этой задаче вам предстоит иллюстрировать этот факт, посчитав реальный уровень доверия для интервалов.

**Реальный уровень доверия (оценка доли покрытия интервалом)**  - доля случаев попадания истинного значения параметра в доверительный интервал.

*Пример:* вы проводите 100 экспериментов (создаете 100 выборок), для каждой выборки рассчитываете АДИ для параметра $\theta$. Истинный параметр попадает в 93 интервала из 100, значит реальный уровень доверия в этом случае будет 0.93.

**Важно:** при такой оценке реального уровня доверия вы используете метод Монте-Карло. Погрешность этого метода составляет $\sim \frac{1}{\sqrt{n}}$, где $n$ - количество выборок, по которым осуществляется оценка. 

*Вопрос:* какое $n$ нужно брать, если вы хотите оценить реальный уровень доверия с точностью до 2 знаков ($\delta = 0.01$)? 

**Ответ:** n = 10000


Сгенерируйте набор выборок из нормального распределения $\mathcal{N}(\theta, 1)$ при $\theta=0$

```python
theta = 0

sample_size = 300
sample_count = 10000

X = np.random.normal(loc=theta, scale=1, size=(sample_count, sample_size))
X.shape
```

Рассмотрим $\hat{\theta} = \overline{X}$.

Напоминание из теории:
$$\hat{\theta} - \frac{z_{\frac{1+\alpha}{2}}\hat{\sigma}}{\sqrt{n}} < \theta < \hat{\theta} + \frac{z_{\frac{1+\alpha}{2}}\hat{\sigma}}{\sqrt{n}}$$

Посчитайте $z$ (используйте функцию `.ppf`)

```python
theta = 0

sample_size = 300

alpha = 0.95

z = sps.norm.ppf((1 + alpha)/ 2)


error = z / np.sqrt(sample_size)

lower_bound_treory = theta - error
upper_bound_treory = theta + error

print(f"Доверительный интервал: ({lower_bound_treory:.4f}, {upper_bound_treory:.4f})")
```

Постройте график зависимости реального уровня доверия от размера выборки.

```python
theta = 0
sample_size = 1000
sample_count = 10000

X = np.random.normal(loc=theta, scale=1, size=(sample_count, sample_size))


sample_sizes = np.arange(10, sample_size+1, 10)
coverage_probabilities = []

for size in sample_sizes:
    sample_means = np.mean(X[:, :size], axis=1)
    sample_std = np.std(sample_means, ddof=1)
    lower_bound = sample_means - z * sample_std / np.sqrt(size)
    upper_bound = sample_means + z * sample_std / np.sqrt(size)
    actual_coverage = np.mean((lower_bound < theta) & (theta < upper_bound))
    coverage_probabilities.append(actual_coverage)

plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, coverage_probabilities, marker='o', linestyle='', color='b', label='real')



plt.xlabel('размер выборки')
plt.ylabel('уровень доверия')
plt.title('Фактическая вероятность покрытия от размера выборки')

x = np.linspace(10, sample_size+1, 1000)
plt.plot(x, sps.norm.ppf((1 + alpha)/ 2) / np.sqrt(x), label="treory")
plt.legend()
plt.grid(True)
plt.show()

```
